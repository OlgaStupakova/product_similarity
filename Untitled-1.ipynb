```python
# to change data types
def change_data_types(df):
    def extract_numeric(val):
        try:
            numeric_part = re.search(r'(\d+)', str(val))
            if numeric_part:
                return int(numeric_part.group())
            else:
                return None
        except ValueError:
            return None

    if 'val_id' in df.columns:
        df['val_id'] = df['val_id'].apply(extract_numeric)
    if 'base_id' in df.columns:
        df['base_id'] = df['base_id'].apply(extract_numeric)

    return df
```

```python
# distribution graphs of column values in a DataFrame
import matplotlib.pyplot as plt

def distribution_graphs(dataframe, color='blue', rows=12, columns=6):
    
    fig, axs = plt.subplots(rows, columns, figsize=(columns * 4, rows * 4))
    
    j_increment = 0  
    
    
    for i in range(rows):
        for j in range(columns):
            
            column = str(i + j + j_increment)
            
            
            data = dataframe[column].value_counts().to_frame().reset_index().sort_values(by=column)
            
            
            axs[i, j].plot(data['index'], data[column], marker='.', linestyle='', color=color)
            axs[i, j].set_title('Feature #' + column + ' Values')
        
        
        j_increment += 5
    
    plt.show()



```

```python
# scaling and displaying scaled vectors
from sklearn.preprocessing import StandardScaler

def scale_datasets(validation_df, base_df, train_df):
  
    scaler = StandardScaler()
    scaler.fit(base_df.iloc[:, 1:])  

    
    scaled_validation_df = validation_df.copy()
    scaled_validation_df.iloc[:, 1:] = scaler.transform(scaled_validation_df.iloc[:, 1:])
    
    
    scaled_base_df = base_df.copy()
    scaled_base_df.iloc[:, 1:] = scaler.transform(scaled_base_df.iloc[:, 1:])
    
    
    scaled_train_df = train_df.copy()
    scaled_train_df.iloc[:, 1:-1] = scaler.transform(scaled_train_df.iloc[:, 1:-1])
    
    return scaled_validation_df, scaled_base_df, scaled_train_df


```

```python
scaled_validation_df, scaled_base_df, scaled_train_df = scale_datasets(validation_df, base_df, train_df)
```
